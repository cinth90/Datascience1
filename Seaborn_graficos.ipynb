Parte 1: NumPy

```python
import numpy as np
# Crear un vector desde una lista
v = np.array([1, 2, 3, 4])
print(f"Esto es v:{v}")

a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

# Suma elemento a elemento
print(f"La suma de a:{a} + b:{b} es: {a+b}")

# Multiplicación escalar
print(f'Esto es a * 2: {a*2}')

# Potencia
print(f'Esto es el cuadrado de b {b ** 2}')
```

```python
# Matriz 2x2
A = np.array([[1, 2], [3, 4]])

# Transpuesta
print(f'A es {A} y traspuesta es {A.T}')

# Inversa: es una matriz que, cuando se multiplica por la matriz original A, da como resultado la matriz identidad (I)
print(f' A es {A} \n y su inversa es {np.linalg.inv(A)}')

# Producto matricial: Para obtener cada elemento de la matriz resultante, se multiplica cada fila de la primera matriz por cada columna de la segunda matriz.
# Se define solo si el número de columnas de la primera matriz es igual al número de filas de la segunda matriz.
B = np.array([[5, 6], [7, 8]])
print(f'El producto matricial de a*b es {np.dot(A, B)}')
```

Mini Desafío: Crear una clase Alumno con métodos para guardar calificaciones y calcular el promedio.

```python
class Alumno:
  def __init__(self, nombre, nota):
    self.nombre = nombre
    self.nota = nota

  def promedio(self):
    promedio = sum(self.nota)/len(self.nota)
    print(f"El promedio del alumno {self.nombre} es {promedio}")
    return sum(self.nota)/len(self.nota)
```

```python
alumno1 = Alumno("juan", [5,7,8,8,8,9,9,9])
alumno1.promedio()
alumno2= Alumno("Ana", [9, 8, 10])
alumno2.promedio()
```

Seaborn para graficos en python


```python
import pandas as pd
import seaborn as sns
import numpy as np #para crear numeros aleatorios
import matplotlib.pyplot as plt

#creando un dataframe a partir de una lista y numeros random
categoria_prod=["Crema", "Perfume","Loción para después de bañarse"]
ventas = np.random.randint(0,50000,3)

dataframe = pd.DataFrame({"Categoria":categoria_prod,"Ventas": ventas})
print(dataframe)

#graficando con sns
sns.set(palette="rocket")
sns.barplot(x="Categoria", y = "Ventas", data=dataframe)

plt.title("Ventas por categoria")
plt.xlabel("Categoria cosmetica")
plt.ylabel("Ventas totales")
plt.figure(figsize=(10,6))

plt.show()
```

Analisis de disttribuciones con Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#cargo los datos con sns
penguins = sns.load_dataset("penguins")
penguins.head()
```

Grafico de cajas


```python
#creando grafico de cajas con sns
plt.figure(figsize=(10,6))
sns.boxplot(x='species', y='flipper_length_mm', data=penguins)
plt.xlabel('Especie')
plt.ylabel('Longitud de las aletas (mm)')
plt.title('Distribucion de la longitud de aletas por especie')
plt.show()
#medio de la caja = mediana
#extremo de la caja = rango intercuartilico
#circulo afuera = Los outliers o datos atípicos, son observaciones que se desvían significativamente del resto de los datos en un conjunto
#bigote = extension de los datos
```

```python
#creando grafico de violin
plt.figure(figsize=(10,6))
sns.violinplot(x='species', y = 'flipper_length_mm', data=penguins, inner='quartile') #quartile agrega las lineas en el int
plt.xlabel('Especie')
plt.ylabel('Longitud de las aletas (mm)')
plt.title('Distribucion  y densidad de la longitud de las aletas por espercie')
plt.show()
#lineas en el interior del grafico representa los cuartiles
```

Comparacion de graficos


```python
#Crea graficos lado a lado para la comparacion
fig, axs = plt.subplots(1, 2, figsize=(15,6))


#grafico 1 = Grafico de cajas
sns.boxplot(ax=axs[0], x='species', y= 'flipper_length_mm', data=penguins)
axs[0].set_title('Grafico de caja')

#grafico 2 = Grafico de violin
sns.violinplot(ax=axs[1], x= 'species', y= 'flipper_length_mm', data=penguins, inner='quartile')
axs[1].set_title('Grafico de violin')

plt.show()
```

Seaborn: Ejemplo practico

```python
#cargando el data set

tips = sns.load_dataset('tips')
tips.head() #muestra los primeros 5 datos del dataset

#creando grafico lado a lado para la comparacion
fig, axs = plt.subplots(1, 2, figsize=(15,6))

sns.boxplot(ax=axs[0], x='day', y='total_bill', data=tips)
axs[0].set_title('Grafico de caja')

sns.violinplot(ax=axs[1], x='day', y='total_bill', data=tips, inner='quartile')
axs[1].set_title('Grafico de violin')

plt.show()
```

Personalizacion de graficos en seaborn

```python
import seaborn as sns

# Establecer el estilo y la paleta
sns.set_style("darkgrid")
sns.set_palette("bright")

# Crear un gráfico de dispersión con etiquetas personalizadas
ax = sns.scatterplot(x="total_bill", y="tip", hue="day", data=tips)
ax.set(title="Relación entre Total de la Cuenta y Propina", xlabel="Total de la Cuenta ($)", ylabel="Propina ($)")

# Personalizar el tamaño de la leyenda y la posición del título
ax.legend(title="Día de la Semana", fontsize=12)
ax.title.set_position([0.5, 1.05])

sns.despine()  # Eliminar bordes superiores y derechos del gráfico
```
